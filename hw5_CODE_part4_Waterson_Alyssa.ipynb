```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
```

```python
R = 1
L= 1
w = np.arange(0,2*np.pi*20,np.pi/4)
h = R/(R**2 + L**2 * w**2)**(1/2)
plt.plot(w,h)
plt.xlabel('w')
plt.ylabel('H(w)')
```

By the way this graph looks, it's for real a low-pass filter. We can see H(w) has a value of 1 at w=0 and H approaches 0 as w gets larger.
