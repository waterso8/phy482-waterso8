```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
```

```python
## First line: EoT/EoI. Second line: EoR/EoI for T and R, all versus theta I
thetaI = np.arange(0,np.pi/2, np.pi/16)
thetaT = .8
mu1 = 1
mu2 = 2
v1 = 2
v2 = 1
rho = np.cos(thetaT)/np.cos(thetaI)
beta = (mu1*v1)/(mu2*v2)
#rhobeta = np.sqrt(beta**2 - np.sin(thetaI)**2)/np.cos(thetaI)
#T = rho*beta*(2/(rho+beta))
#R = ((1-rho*beta)**2)/((1+rho*beta)**2)
#plt.plot(thetaI, T)
#plt.plot(thetaI, R)
```

```python
x = 0*thetaI
EorEoI = (1-(rho*beta))/(1+(rho*beta))
EoTEoI = (2/(1+(rho*beta)))
plt.plot(thetaI, EoTEoI, label = 'EoT/EoI', color = 'g')
plt.plot(thetaI, -EorEoI, label = 'EoR/EoI', color = 'r')
plt.plot(thetaI, x, label = 'the 0 axis', color = 'black')
plt.xlabel("theta-I (in radians)")
plt.ylabel("E/E")
plt.title("E/E VS Theta-I")
plt.legend()
```

You can make this graph more interpretable if you make v1 a bit larger than v2, but as requested, n2/n1 (= v1/v2) is 2.

```python
T = rho*beta*EoTEoI**2
R = EorEoI**2
plt.plot(thetaI, T, label = 'T', color = 'g')
plt.plot(thetaI, R, label = 'R', color = 'r')
plt.legend()
TR = T + R
plt.plot(thetaI, TR, label = 'T + R', color = 'purple')
print(TR[0])
plt.legend()
```
