```python
import matplotlib.pyplot as plt
% matplotlib inline
import math
import numpy as np
```

## Part 1

```python
x = np.linspace(-.01,.01,10)
y = np.linspace(-.01,.01,10)

X, Y = np.meshgrid(x,y)
E = X/(Y**2+X**2)**(3/2) 
F = Y/(Y**2+X**2)**(3/2)
plt.quiver(X, Y, E, F)
```

## Part 2

```python
x = np.linspace(-1,1)
y = np.linspace(-1,1)

X, Y = np.meshgrid(x,y)
E = 2*X/(Y**2+X**2)**(5/2) 
F = -Y/(Y**2+X**2)**(5/2)

plt.axis([-.25,.25,-.25,.25])
plt.quiver(X, Y, E, F)
```
