```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
w = 1
L = 1
R = 1
C = 1
V_0 = 1
t = np.arange(1,100,.1)
I = V_0*np.cos(w*t) * ((1/R**2) + (w*C - (1/(w*L)))**2)**(1/2)
plt.plot(t,I)
```

```python
t = 1
w = np.linspace(0,2*np.pi,990)
plt.plot(w, I)
```
