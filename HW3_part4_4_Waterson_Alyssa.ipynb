```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = []
for i in range(0,100): #giving a changing alpha value
    alph = 10^i
    a.append(alph)
    
initial_B = 1 
t = np.linspace(0,100,100) #defining the amount of seconds
x = np.linspace(1,2,100) # defining how many spaces to step
v = 1/100 #taking one step for each 100 expected
for i in a: #running through each alpha to get the function
    i = 0 # if i did not initialize this, it looked like the function spaced out
    emf = (i * t**2 + initial_B * t) * v + x * (2 * i * t + initial_B)
```

```python
plt.plot(emf, a)
```
